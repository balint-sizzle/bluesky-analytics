```python
from atproto import Client
import json
import requests
import dotenv
import os

dotenv.load_dotenv()

email = os.getenv("BLUESKY_EMAIL")
password = os.getenv("BLUESKY_PASSWORD")

client = Client()
client.login(email, password)
```

```python
res = client.get_author_feed("profgalloway.com", limit=10).model_dump()
# j = json.loads(res)
with open("author_feed.json", "w", encoding="utf-8") as f:
    json.dump(res, f, ensure_ascii=False, indent=4, )
# dir(res)
```
